```python
#Remove null values
df.dropna(inplace=True)
```

```python
#Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
#Convert dates
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')
```

```python
#Extract date components
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Weekday'] = df['Order Date'].dt.day_name()
```

```python
#Create calculated fields
df['Delivery Time (Days)'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Profit Margin'] = df['Profit'] / df['Sales']
```

```python
df
```

```python
#Delayed Deliveries vs High Sales
# Average delivery time per region
delay_sales = df.groupby('Region').agg({
    'Delivery Time (Days)': 'mean',
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()
```

```python
print("Average delivery time per region:\n",delay_sales)
```

```python
# Flag regions with delivery time > average but high sales
average_delivery = df['Delivery Time (Days)'].mean()
print("Average delivery time:",average_delivery)
average_sales = df['Sales'].mean()
print("Average Sales:",average_sales)
```

```python
delay_sales['Delivery Flag'] = delay_sales['Delivery Time (Days)'].apply(lambda x: 'Delayed' if x > average_delivery else 'On Time')
delay_sales['Sales Flag'] = delay_sales['Sales'].apply(lambda x: 'High Sales' if x > average_sales else 'Low Sales')
```

```python
delay_sales.head()
```

```python
# Filter regions with delay but high sales
inefficient_regions = delay_sales[(delay_sales['Delivery Flag'] == 'Delayed') & (delay_sales['Sales Flag'] == 'High Sales')]

print("Regions with delayed deliveries but high sales:\n",inefficient_regions)
```

```python
# Save the cleaned and updated dataset to a new CSV file
df.to_csv("Cleaned_Superstore_Data.csv", index=False)

# Also save the unique insight data (inefficient regions)
inefficient_regions.to_csv("Regions_Delayed_High_Sales.csv", index=False)

print("Cleaned dataset and insight data saved successfully!")
```
